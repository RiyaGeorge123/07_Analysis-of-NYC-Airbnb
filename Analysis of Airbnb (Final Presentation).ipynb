```python
dataset = pd.read_csv("AB_NYC_2019.csv")# importing csv datafile
```

```python
dataset.drop(['id','name','host_id','host_name','neighbourhood','latitude','longitude','number_of_reviews','last_review','reviews_per_month','calculated_host_listings_count','availability_365'], axis=1, inplace=True)
# Drop Columns we didn't need 
```

```python
dataset.head() #Showing remaning columns after drop
```

Encoding neighbourhood_group and room_type to have integer values starting from 0   

```python
def Encode(dataset):
    for column in dataset.columns[dataset.columns.isin(['neighbourhood_group','room_type'])]:
        dataset[column] = dataset[column].factorize()[0]
    return dataset

dataset_new = Encode(dataset.copy())
# renaming dataset to dataset_new to represent encoded values
```

```python
#dataset_new
# 0 = Brooklyn, 1 = Manhattan, 2 = Queens, 3 = Staten Island, 4 = Bronx
# 0 = Private Room, 1 = Entire home/apt 2 = Shared room
```

```python
dataset_new.head() #showing first 5 rows of dataset_new 
```

##Model Chosen LinearRegression Predicitve Modeling 

```python
from sklearn.linear_model import LinearRegression # importing linear regression model
from sklearn.model_selection import train_test_split # importing train test split 
from sklearn.metrics import r2_score # importing r2_score
```

```python
from sklearn import metrics

def print_evaluate(true,predicted):
    
    r2_square = metrics.r2_score(true, predicted)
    
    print('\nR2 Square', r2_square)
    #Showing the strength of the relationship between 
    #independent variable and dependent variable
```

```python
import numpy as np #mathematical and logical operations on arrays can be performed
```

*X is our independent variable( train data) used to predict the outcome. In our example, neighbourhood_group serves as our independent variable.
*Y is our dependent variable test data used to measure the performance. In our example price is our dependent variable. The price of Airbnb is dependent on the neighbourhood_group and other independent variables.   

```python
X = dataset_new[['neighbourhood_group']]
y = dataset_new.price
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression(normalize=True)
lin_reg.fit(X_train,y_train)
```

```python
lin_reg.fit(X_train,y_train)
```

```python
y_pred = lin_reg.predict(X_test)
```

```python
print("Error/Accuracy Analysis :- ")
print_evaluate(y_test,y_pred)
```

```python
def Final_Price_Check(y_test,y_pred,):
    dataset = {"Actual Price" : y_test.values, "Predicted Price" : y_pred} #Our y_pred will represent our predicted price
    result = pd.DataFrame(dataset)
    return result
```

```python
Final_Price_Check(y_test,y_pred).head(10)
```
